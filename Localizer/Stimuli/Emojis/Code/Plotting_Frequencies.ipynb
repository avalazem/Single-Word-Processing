```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

emoji_df = pd.read_csv(r"C:\Users\ali_a\Desktop\Single_Word_Processing_Stage\Single_Word_Processing\Localizer\Stimuli\Emojis\frequencies_w_emoji.txt", sep='\t', header=None) 
emoji_df.columns = ['Emoji', 'Frequency']
emoji_df['Index'] = range(1, len(emoji_df) + 1)
print(emoji_df.head())
plt.figure(figsize=(12, 5))
sns.histplot(data=emoji_df, x='Index', y='Frequency', log_scale=True)
plt.title("Emoji Frequency Distribution (Log Scale)", fontsize=12)
plt.xlabel('Emoji')
plt.xticks([])
plt.show()
```

```python
# Plotting the data
plt.figure(figsize=(12, 5))
sns.histplot(data=emoji_df, x='Emoji', y='Frequency', log_scale=True)
plt.title("Emoji Frequency Distribution (Log Scale)", fontsize=12)
plt.xlabel('Emoji')

# Setting x-axis ticks to display every 20th emoji
xticks = emoji_df['Emoji'][::20]
plt.xticks(ticks=range(0, len(emoji_df), 20), labels=xticks)

plt.show()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the grayscale image
img = Image.open(r"C:\Users\ali_a\Desktop\Single_Word_Processing_Stage\Single_Word_Processing\Localizer\Stimuli\Emojis\Emoji Localizer png's\yum_no_mask.png").convert("RGB")  # Ensure RGB format

# Convert image to a NumPy array and flatten it to 1D
pixels = np.array(img).flatten()

# Create a single X-axis value repeated for all pixels
x_axis = np.ones_like(pixels)  # Same length as pixels, but all set to 1

# Scatter plot
plt.figure(figsize=(4, 6))
plt.scatter(x_axis, pixels, color="black", alpha=0.5, s=1)  # Small dots for visibility
plt.ylabel("Pixel Intensity (0-255)")
plt.xticks([])  # Hide X-axis ticks
plt.title("Pixel Intensity Distribution")

plt.show()
```
